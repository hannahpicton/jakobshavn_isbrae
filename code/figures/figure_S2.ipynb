### Figure S2 - ITS_LIVE Velocity at T, T5, T10, T15, T20 & T30.


```python
# IMPORT MODULES 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import numpy as np 
import datetime

# OPEN THE ITS_LIVE 6-DAY PAIRS CSV FILE 
ITS_LIVE_data = pd.read_csv(r'R:/JAKOBSHAVN/CODE/github/jakobshavn_isbrae/data/its_live_sampled_6_12_day.csv')

# DEFINE THE VELOCITY VARIABLE FOR EACH POINT
POINT1_VELOCITY = ITS_LIVE_data['POINT_1_VELOCITY']
POINT2_VELOCITY = ITS_LIVE_data['POINT_2_VELOCITY']
POINT3_VELOCITY = ITS_LIVE_data['POINT_3_VELOCITY']
POINT4_VELOCITY = ITS_LIVE_data['POINT_4_VELOCITY']
POINT5_VELOCITY = ITS_LIVE_data['POINT_5_VELOCITY']
POINT6_VELOCITY = ITS_LIVE_data['POINT_6_VELOCITY']
POINT7_VELOCITY = ITS_LIVE_data['POINT_7_VELOCITY']

# DEFINE THE MID-DATE VARIABLE
ITS_LIVE_data['mid_date'] = pd.to_datetime(ITS_LIVE_data['mid_date'], format='%Y/%m/%d')
```

```python
# PRODUCE A FIGURE WITH EACH SAMPLE POINT SHOWN ON AN INDIVIDUAL SUB-PLOT
plt.rcParams["font.family"] = "Arial"
fig, axes = plt.subplots(3, 2, figsize=(10, 8), sharex=True)
plt.subplots_adjust(hspace=0.1)
scatter_colors = ['#a30b37', '#f77f00', '#1c7c54', '#219ebc', '#10678e', '#023047']
labels = ['T', 'T5', 'T10', 'T15', 'T20', 'T30']

for i, ax in enumerate(axes.flatten()):
    year_locator = mdates.YearLocator(base=1, month=1, day=1)
    ax.xaxis.set_major_locator(year_locator)
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
    min_x_value = pd.to_datetime('2018-01-01', format='%Y-%m-%d')
    max_x_value = pd.to_datetime('2023-01-01', format='%Y-%m-%d')
    ax.grid(axis = 'x', alpha=0.5, linestyle='-', color='black')
    ax.xaxis.grid(True, which='minor', linestyle='--', alpha=0.3)
    three_month_intervals = mdates.MonthLocator(interval=3)
    ax.xaxis.set_minor_locator(three_month_intervals)
    plt.xlim(min_x_value, max_x_value)
    ax.text(1, 1, labels[i], color=scatter_colors[i], transform=ax.transAxes, ha='right', va='top', fontsize=14, bbox=dict(facecolor='white', edgecolor='white', boxstyle='round,pad=0.3'), weight='bold')
    if i // 2 == 2:
        ax.set_xlabel('Year', fontsize=13, labelpad=10)

axes[0, 0].scatter(ITS_LIVE_data['mid_date'], POINT1_VELOCITY/1000, color='#a30b37', s=3, marker='o', alpha=0.85, label='T', zorder=20)
axes[0, 1].scatter(ITS_LIVE_data['mid_date'], POINT2_VELOCITY/1000, color='#f77f00', s=3, marker='o', alpha=0.85, label='T5', zorder=20)
axes[1, 0].scatter(ITS_LIVE_data['mid_date'], POINT3_VELOCITY/1000, color='#1c7c54', s=3, marker='o', alpha=0.85, label='T10', zorder=20)
axes[1, 1].scatter(ITS_LIVE_data['mid_date'], POINT4_VELOCITY/1000, color='#219ebc', s=3, marker='o', alpha=0.85, label='T15', zorder=20)
axes[2, 0].scatter(ITS_LIVE_data['mid_date'], POINT5_VELOCITY/1000, color='#10678e', s=3, marker='o', alpha=0.85, label='T20', zorder=20)
axes[2, 1].scatter(ITS_LIVE_data['mid_date'], POINT7_VELOCITY/1000, color='#023047', s=3, marker='o', alpha=0.85, label='T30', zorder=20)

axes[0, 0].set_ylim(5.5, 18.5)
axes[0, 1].set_ylim(4.5, 11.5)
axes[1, 0].set_ylim(3.5, 8.5)
axes[1, 1].set_ylim(2.8, 5.2)
axes[2, 0].set_ylim(2.25, 4.65)
axes[2, 1].set_ylim(0.55, 1.25)

fig.text(0, 0.5, 'Ice Velocity (km/yr)', va='center', rotation='vertical', fontsize=13)
plt.tight_layout(rect=[0.01, 0.07, 1, 0.95])
#plt.savefig('R:/JAKOBSHAVN/CODE/github/jakobshavn_isbrae/figures/figure_S2.png', dpi=300)
```
