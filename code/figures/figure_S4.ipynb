### Figure S4
Ice surface elevation sampled along the centreline of a nearby land-terminating margin using ArcticDEM 2m strips. 


```python
# IMPORT REQUIRED MODULES 
import os
import numpy as np
import pandas as pd
import geopandas as gpd
import rioxarray as rxr
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from datetime import datetime
```

```python
# Load the CSV file containing the sampled ice surface elevation data
elevation_data = pd.read_csv('R:/JAKOBSHAVN/CODE/github/jakobshavn_isbrae/data/ice_elevation_ArcticDEM/land_terminating_ArcticDEM_2M.csv')
elevation_data.set_index('distance_from_T', inplace=True)

# Select the 'June' profiles for each year
selected_columns = ['20180627', '20190605', '20200519', '20210604']
elevation_data_june = elevation_data[selected_columns]

# Specify colors for each profile if available
profile_colors = {'20180627': '#023047',  
                  '20190605': '#219ebc',  
                  '20200519': '#83c5be',  
                  '20210604': '#a7c957'}
```

```python
# PLOT FIGURE S4A # 
plt.rcParams["font.family"] = "Arial"
fig, ax = plt.subplots(figsize=(8, 4))
plt.grid(axis='x', alpha=0.2, linestyle='-', color='black')
plt.xlabel('Distance along profile (km)', fontsize=12, labelpad=10)
plt.ylabel('Elevation (m)', fontsize=12, labelpad=10)

for column in elevation_data_june.columns:
    color = profile_colors.get(column, 'gray')  # Default to gray if color not specified
    date_formatted = datetime.strptime(column, '%Y%m%d').strftime('%Y-%m-%d')
    plt.plot(elevation_data_june.index / 500, elevation_data_june[column], label=date_formatted, color=color, linewidth=1.25)

x_ticks = np.arange(0, 4.5 + 0.5, 0.5)  # Adjust the range as needed
plt.xticks(x_ticks, [f"{x:.1f}" for x in x_ticks], fontsize=12)  # Formatting ticks to one decimal place
plt.yticks(fontsize=12)
plt.xlim(0, 4.2)
plt.ylim(75, 425)
plt.legend(fontsize=11, loc='upper left', bbox_to_anchor=(0.02, 0.98), framealpha=0.9, edgecolor='gray', fancybox=True, markerscale=2, handlelength=1.5, borderpad=0.5, labelspacing=0.4)
plt.tight_layout()
plt.show()
#plt.savefig('R:/JAKOBSHAVN/CODE/github/jakobshavn_isbrae/figures/figure_S4a.png', dpi=300)
```
